```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
import locale
```

```python
locale.setlocale(locale.LC_ALL, 'deu_DEU')

file = "./test.txt"
data=np.loadtxt(file)
```

```python
fig = plt.figure(figsize=(10, 10))
fig.suptitle("Ruchy browna")

ax = plt.axes((0, 1000), (0, 1000))
point, = ax.plot([], [])
plt.close()

def animate(i):
    point.set_data(data[i,0], data[i,1])
    return point,

anim = animation.FuncAnimation(fig, animate, frames=400, interval=10, blit=True)
HTML(anim.to_html5_video())   
```
